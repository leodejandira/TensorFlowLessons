```python

import pandas as pd
import plotly.io as pio

df = pd.read_csv('/content/drive/MyDrive/Astrofísica/Google Colab/LUCS_1201_data.csv')

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go

import numpy as np

init_notebook_mode(connected=True)

x = df['x']
y = df['y']
z = df['z']

splot = go.Scatter3d(
    x=np.array(x),
    y=np.array(y),
    z=np.array(z),
    mode='markers',
    marker=dict(
        color='rgb(255, 0, 0)',
        size=4
    ),
    opacity=0.6
)

pdata = [splot]
layout = go.Layout(
             scene=dict(
                 aspectmode='data'
         ))
fig = go.Figure(data=pdata, layout=layout)
iplot(fig, image='svg', filename='scatterplot.svg', image_width=600, image_height=600)



fig.write_image("sample_data/Lucs_1201-static.png")
```
